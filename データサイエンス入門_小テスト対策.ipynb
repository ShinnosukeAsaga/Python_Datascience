# 1.区間推定(正規分布)
ある調査で大学生300人に、1週間にビールを飲んだ回数を
聞いたところ、平均が2.4回、標準偏差が1.1回であった。
信頼度95％で大学生の平均飲用回数の区間を推定しよう。

````python
#問題によって適宜変更する　←重要

from scipy.stats import norm
student=300 #標本数
ave=2.4 #平均
sd=1.1  #標準偏差

# 標準正規分布の累積分布関数 (CDF) の逆関数 (PPF) を使用してZスコアを計算
z_score = norm.ppf(1 - (1 - 0.95) / 2) #信頼度 95%（1-0.95=0.05）

# 標準誤差を計算
standard_error = sd / np.sqrt(student)

# 信頼区間を計算
lower_bound = ave - z_score * standard_error
upper_bound = ave + z_score * standard_error

print(f"信頼度95%での大学生の平均飲用回数の区間は [{lower_bound:.2f}, {upper_bound:.2f}] 回です。")
````

# 2. 区間推定(t検定)
A社サラダ油ボトルの重さは正規分布に従うとし、800g入りの ボトル9本を選んで、重さを測定したところ、

その値は {807, 811, 801, 798, 798, 795, 803, 805, 804}g でした。 このデータを用いて、母平均の99％の信頼区間を求めてください。


````python
import numpy as np
from scipy.stats import t
# 与えられたデータ　(適宜変更)
data = np.array([807, 811, 801, 798, 798, 795, 803, 805, 804])

# 標本サイズ
n = len(data)

# 標本平均を計算
sample_mean = np.mean(data)

# 標本標準偏差を計算 (不偏標準偏差を使用するため ddof=1 を指定)
sample_std = np.std(data, ddof=1)

# 自由度
degrees_of_freedom = n - 1

# 累積確率が 1 - 0.005 = 0.995 となるt値を t.ppf() で求める
t_score = t.ppf(1 - (1 - 0.99) / 2, degrees_of_freedom)

# 標準誤差を計算
standard_error = sample_std / np.sqrt(n)

# 信頼区間を計算
lower_bound = sample_mean - t_score * standard_error
upper_bound = sample_mean + t_score * standard_error

print(f"母平均の99%信頼区間は [{lower_bound:.2f}, {upper_bound:.2f}] g です。")
````

# 3.区間推定($χ^2$検定)

Aさんは化学実験である溶液のpHを測定し、下のデータを得た。
この溶液のpH測定値について、母分散の95％信頼区間を求めなさい。(小数点第3位まで)

|番号|pH値|
|:--:|:--:|
|1|7.68|
|2|7.48|
|3|7.28|
|4|7.98|
|5|7.09|

````python
import numpy as np
from scipy.stats import chi2
# 与えられたデータ
ph_values = np.array([7.68, 7.48, 7.28, 7.98, 7.09])

# 標本サイズ
n = len(ph_values)

# 標本平均を計算
sample_mean_ph = np.mean(ph_values)

# 標本分散を計算 (不偏分散を使用するため ddof=1 を指定)
sample_variance = np.var(ph_values, ddof=1)
# 自由度
degrees_of_freedom_ph = n - 1

# 信頼度95%の場合、両側確率が5%なので、片側確率は2.5% = 0.025
# 下側カイ二乗値 (累積確率 0.025) と 上側カイ二乗値 (累積確率 0.975) を chi2.ppf() で求める
chi2_lower = chi2.ppf((1 - 0.95) / 2, degrees_of_freedom_ph)
chi2_upper = chi2.ppf(1 - (1 - 0.95) / 2, degrees_of_freedom_ph)

# 母分散の信頼区間を計算
lower_bound_var = (degrees_of_freedom_ph * sample_variance) / chi2_upper
upper_bound_var = (degrees_of_freedom_ph * sample_variance) / chi2_lower

print(f"母分散の95%信頼区間は [{lower_bound_var:.3f}, {upper_bound_var:.4f}] です。")
````

# 4.P検定(仮説検定)

ある工場での経験によると 7mmボルトの規格の製品の寸法はほぼ正規分布をしており，その標準偏差は0.20mmであるという． ある日の製品から100個のボルトをランダムに抽出したところ，その寸法の平均が7.09mmであった． この日の製品寸法の平均は規格から外れているか、有意水準５％で検定せよ．

````python
import math

# 帰無仮説 H0: 母平均 μ = 7.0 mm
mu_0 = 7.0       # 帰無仮説での母平均
sigma = 0.20     # 母標準偏差 (既知)
n = 100          # 標本サイズ
x_bar = 7.09     # 標本平均
alpha = 0.05     # 有意水準

# 検定統計量 Z  = (標本平均 - 母平均) / (母標準偏差 / sqrt(標本サイズ))
z_stat = (x_bar - mu_0) / (sigma / math.sqrt(n))

# p値の計算 (両側検定)
# 標準正規分布の累積分布関数 (CDF) を使う。
p_value = 2 * (1 - 0.5 * (1 + math.erf(abs(z_stat) / math.sqrt(2))))


# 検定結果
if p_value < alpha:
    print(f"有意水準 {alpha} で帰無仮説を棄却する。")
else:
    print(f"有意水準 {alpha} で帰無仮説を棄却しない。")
````

# 5. 仮説検定($χ^2$検定)
H社で製造中のリベットの直径の分散は0.42とバラツキが大き
いので、新しい製造方法を導入した.
この新しい製造方法で作られたリベットの中から10個のサン
プルを無作為抽出し、直径を測定したところ, 次の結果になました.
```
 ｛35.2, 34.5,34.9,35.2,34.8; 35.1, 34.9,35.2,34.9,34.8｝
 ```
新しい製造方法で、リベットの直径のバラツキは改善されたか、
有意水準5％で検定せよ.

````python
import numpy as np
from scipy.stats import chi2

# 標本データ
data = np.array([35.2, 34.5, 34.9, 35.2, 34.8, 35.1, 34.9, 35.2, 34.9, 34.8])

sigma2_0 = 0.4**2  # 帰無仮説での母分散 (古い製造方法での分散)
n = len(data)      # 標本サイズ
alpha = 0.05       # 有意水準

# 標本分散の計算 (不偏分散)
sample_variance = np.var(data, ddof=1)

# 検定統計量 χ^2 の計算
# χ^2 = (n-1) * 標本分散 / 母分散
chi2_stat = (n - 1) * sample_variance / sigma2_0

# p値の計算 (左側検定)
p_value = chi2.cdf(chi2_stat, df=n-1)

print(f"χ^2 統計量: {chi2_stat:.4f}")
print(f"p値: {p_value:.4f}")
#臨界値
critical_value = chi2.ppf(alpha, df=n-1)
print(f"有意水準 {alpha} で棄却する際の臨界値: {critical_value:.4f}")
# 検定結果
if p_value < alpha:
    print(f"有意水準 {alpha} で帰無仮説を棄却する。")
else:
    print(f"有意水準 {alpha} で帰無仮説を棄却しない。")
````
